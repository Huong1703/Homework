1. Đọc 1 file và in ra màn hình 5 dòng đầu tiên, nếu file không đủ 5 dòng thì in toàn bộ nội dung file

```python
def docFile():
    with open('123.txt', 'r', encoding='utf-8') as file:
        lines = file.readlines()

        for line in lines[:5]:
            print(line.strip())

docFile()
```

2. Đọc 1 file và in ra màn hình 5 dòng cuối cufg, nếu file không đủ 5 dòng thì in toàn bộ nội dung file

```python
def docFile():
    with open('123.txt', 'r', encoding='utf-8') as file:
        lines = file.readlines()

        if len(lines) <= 5:
            for line in lines:
                print(line.strip())
        else:
            for line in lines[-5:]:
                print(line.strip())

docFile()
```

3. Đọc 1 file, tìm và in ra nội dung của dòng dài nhất trong file đó

```python
def docFile():
    with open('123.txt', 'r', encoding='utf-8') as file:
        lines = file.readlines()
        longest_line = max(lines, key=len)
        print(longest_line.strip())

docFile()
```

4. Đọc 1 file, tìm và in ra từ dài nhất trong file

```python
def docFile():
    with open('123.txt', 'r', encoding='utf-8') as file:
        content = file.read()
        words = content.split()

        longest_word = max(words, key=len)
        print(longest_word)

docFile()
```

5. Đọc 1 file, thông kê và in ra tất cả các chữ cái có trong file và số lần xuất hiện các chữ số đó

```python
def docFile():
    with open('123.txt', 'r', encoding='utf-8') as file:
        content = file.read()
        letter_counts = {}

        for char in content:
            if char.isalpha():
                char = char.lower()
                letter_counts[char] = letter_counts.get(char, 0) + 1

        for letter, count in letter_counts.items():
            print(f"{letter}: {count}")

docFile()
```

6. Đọc 1 file, thống kê và tần xuất xuất hiện của tất cả các từ trong file, in theo thứ tự giảm dần của tần số xuất hiện

```python
from collections import Counter

def docFile():
    with open('123.txt', 'r', encoding='utf-8') as file:
        content = file.read()
        words = content.split()

        word_counts = Counter(words)
        sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
        for word, frequency in sorted_word_counts:
            print(f"{word}: {frequency}")

docFile()
```

7. Đọc tập tin văn bản abc.txt sau đó chuyển nội dung đọc dược sang chữ in hoa và ghi vào tập tin xyz.txt

8. Đọc tập tin văn bản abc.txt, lọc lấy tất cả các số trong dữ liệu đọc được và ghi chúng lần lượt vào tập tin number.txt, mỗi số ghi trên 1 dòng

```python
with open('123.txt', 'r') as file:
    content = file.read()

numbers = [char for char in content if char.isdigit()]

with open('number.txt', 'w') as file:
    for number in numbers:
        file.write(number + '\n')
```
